```python
ps.sqldf("""

SELECT 
    *
FROM titanic_disaster_dataset

""")

# This is the basic Syntax for writing any SQL code 
# If you notice the " * " (star symbol) is used to get full dataset. In the result below you'll see a small break this is because the User Interface(UI) of jupyter notebook dont want to flood the screen with so much data.
# don't worry the whole dataset is up and running once you write the " * " (star symbol)
```

```python
ps.sqldf("""

SELECT 
    vatsal.* 
FROM titanic_disaster_dataset AS vatsal 

""")


# Sometimes the variable name you give to your data set is really long and its a good practice to give it an alias to it so that its understandble and you dont have to write the whole thing again and again.

# you will understand why we used the word (or alias) "vatsal"(my name) which is small to write and not time consuming.

# an alias can be given by syntax word AS (always in capital).

# This makes things easier to understand for yourself and for anyone new reading your code because it is easy to see from which dataset each column is coming from.
```

```python
ps.sqldf("""

SELECT 
    vatsal.*  
FROM titanic_disaster_dataset AS vatsal
LIMIT 5

""")

# This is inbuilt function similar to .head() as we used earlier which gives top 5 rows.

# you can have as many entries you want by using LIMIT followed by the number of entires you want i.e. 10,20,30 etc.
```

```python
#Taking only Passenger ID and the name from the dataset

ps.sqldf("""

SELECT 
    vatsal.PassengerId,
    vatsal.Name
FROM titanic_disaster_dataset AS vatsal
LIMIT 5

""")

# Now you will understand the use of Alias. Writing the whole ("titanic_disaster_dataset") is too time consuming and makes the code look chunky

# hence the use of alias will make your code look simpler and easy to understand. In this case alias name is ("Vatsal")
```

```python
#Selecting everything from the dataset for passengers where name contains 'Mrs.'

ps.sqldf("""

SELECT 
    df.*
FROM titanic_disaster_dataset AS df
WHERE df.Name LIKE "%MRS.%"
LIMIT 5

""")

# For text/string columns, a powerful way of filtering is using "LIKE" keyword. 
# It works by the specifying the substring you want in a string and surrounding the substring with '%'s. Here is how it looks with examples below -

# For Names starting with "ABC" - Name LIKE "ABC%"
# For Names ending with "ABC" - Name LIKE "%ABC"
# For Names which contain "ABC" - Name LIKE "%ABC%"

# lIKE is not caase sensitive. i.e. it will give results for both upper case and lower case
```

```python
#Concatenating Sex and Ticket

ps.sqldf("""

SELECT 
    df.Sex,
    df.Ticket,
    df.Sex || df.Ticket || df.Age AS Sex_Plus_age_ticket
FROM titanic_disaster_dataset AS df
LIMIT 5    
    
""")

# Concatenating meaning - link (things) together in a chain or series.

# to concatenate use || symbol. (You can also conatenate mulitple items)
```

```python
#Using CASE WHEN to create buckets for age
ps.sqldf("""

SELECT 
    df.Name,
    df.PassengerID,
    df.Age,
    CASE 
        WHEN df.Age IS NULL THEN 'Missing'
        WHEN df.Age <18 THEN 'INFANT'
        ELSE 'ADULT' 
    END AS age_bucket
FROM titanic_disaster_dataset AS df
LIMIT 15  
    
""")

# you can use CASE WHEN whenever you are working with booleans.
# ALWAYS REMEMBER you always must end your booleans with "END AS" or else it will throw error.
```

```python
#Using SUBSTR() to take out a substring

ps.sqldf("""

SELECT 
    df.Name,
    SUBSTR(df.Name,2, 4) AS out
FROM titanic_disaster_dataset AS df
LIMIT 5
    
""")

# SUBSTR() function can be used to take out a part of a string from the text columns in a dataset.
# here from name dataset we take out letters from position 2 to 4 and print the output in new coloumn 
```

```python
#Using LOWER() and UPPER() to convert text to lower and upper case

ps.sqldf("""

SELECT 
    df.Name,
    LOWER(df.Name) AS lcase_name,
    UPPER(df.Name) AS ucase_name
FROM titanic_disaster_dataset AS df
LIMIT 5
    
""")

# this is self explainatory.
```

```python
#Getting all distinct values of "Embarked" column

ps.sqldf("""

SELECT 
    DISTINCT df.Sex
FROM titanic_disaster_dataset AS df
    
""")

# You can get all the unique values in a column using "DISTINCT" keyword
```

```python
#Aggregating information at "Embarked" column level

ps.sqldf("""

SELECT 
    df.Embarked,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age
FROM titanic_disaster_dataset AS df
GROUP BY df.Embarked
""")

# in this we select entries from Embarked dataset and get average, sum etc etc. 
# always remember that after you are done  finish it with "GROUP BY" followed by the dataset name you used.
```

```python
#Aggregating information at "Embarked" and "Sex" levels

ps.sqldf("""

SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df
GROUP BY df.Embarked,
    df.Sex

""")

# Another example but this time we are using multiple Datas.
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions

ps.sqldf("""

SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
FROM titanic_disaster_dataset AS df

WHERE age >= 18
    AND age <=60
    
GROUP BY df.Embarked,
    df.Sex

""")

#where age is between 18 and 60. so it will only give result of people having age between 18 to 60.
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also sorting the output

ps.sqldf("""

SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df

WHERE age >= 18
    AND age <=60
    
GROUP BY df.Embarked,
    df.Sex

ORDER BY df.Embarked,
    df.Sex

""") 

# "ORDER BY" is syntax to sort the coloumn alphabatically or numerically. just like we use in MS Excel.
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also sorting the output in descending order

ps.sqldf("""

SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df

WHERE age >= 18
    AND age <=60
    
GROUP BY df.Embarked,
    df.Sex

ORDER BY df.Embarked DESC,
    df.Sex DESC

""")

# to sort in descending order we use syntax "DESC"
# to sort in ascending order we use syntax "ASC"
```

```python
#Aggregating information at "Embarked" and "Sex" levels 
#and also using "WHERE" to filter basis some conditions
#and also using "HAVING" for fitering the output for buckets where average age is more than 30
#and also using "ORDER BY" for sorting the output in descending order

ps.sqldf("""

SELECT 
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
    
FROM titanic_disaster_dataset AS df

WHERE age >= 18
    AND age <=60
    
GROUP BY df.Embarked,
    df.Sex

HAVING avg_age > 30

ORDER BY df.Embarked DESC,
    df.Sex DESC

""")
```

```python
### DIFFERENCE BEWTEEN WHERE & HAVING (VERY IMPORANT)

# The main difference between them is that the WHERE clause is used to specify a condition for filtering records before any 
# groupings are made.

# the HAVING clause is used to specify a condition for filtering values from a group.
```
